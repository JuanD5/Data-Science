```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('sine wave from')

plt.plot(x,y)
plt.show()
```

#### Bar plots. 

```python
x = [2,8,10]
y = [11,16,9]
x2  = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('Bar Graph')
plt.xlabel(' X axis')
plt.ylabel('Y axis')
```

#### Histograms. 

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) # represents the density or count  of the numbers in the array 
plt.hist(a)
plt.title('histogram')
plt.show()
```

#### Box plots

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert = True, patch_artist= True)
```

#### the last line indicates the 0 percentile , the bottom of the box is the 25% the red line is the 50% percentile the next line the 75% and the final black line is the 100% 

#### Pie chart 

```python
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0.1,0.1,0.1) # how far all the pieces are from the center of the graph.  

plt.pie(sizes,explode= explode, labels= labels, colors= colors, autopct='%1.1f%%',shadow=True) # autopct 1.1 floating format for the percentages. 
plt.axis('equal')
plt.show()
```
